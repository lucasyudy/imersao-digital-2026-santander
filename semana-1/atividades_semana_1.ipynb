```python
# Desafio: Calculadora de Pedidos (Cafeteria)
# Aula 4 - Atividade 2
# Reprodução do algoritmo estudado na Alura

preco_hamburguer = 12
preco_batata = 7
preco_refri = 5

qtd_hamburguer = 2
qtd_batata = 1
qtd_refri = 2

total_hamburguer = qtd_hamburguer * preco_hamburguer
total_batata = qtd_batata * preco_batata
total_refri = qtd_refri * preco_refri

total_pedido = total_hamburguer + total_batata + total_refri

print(f"🍔 Hambúrguer: {qtd_hamburguer} x R$ {preco_hamburguer} = R$ {total_hamburguer}")
print(f"🍟 Batata:      {qtd_batata} x R$ {preco_batata}  = R$ {total_batata}")
print(f"🥤 Refri:       {qtd_refri} x R$ {preco_refri}  = R$ {total_refri}")
print("-" * 40)
print(f"TOTAL DO PEDIDO: R$ {total_pedido}")
```

```python
# Desafio: Conversor de Moedas
# Aula do módulo "Lógica de Programação"
# Algoritmo: conversão USD → BRL

valor_usd = 100.00
cotacao_brl = 5.50
valor_brl = valor_usd * cotacao_brl

print(f"${valor_usd:.2f} USD → R$ {valor_brl:.2f} BRL (cot. R$ {cotacao_brl:.2f})")
```

```python
# Desafio: Classificador de Desempenho Acadêmico
# Módulo: Lógica de Programação
# Simulação: sistema de podmiação baseada em notas

alunos = [
    {"nome": "João",  "nota": 8.5},
    {"nome": "Maria", "nota": 6.0},
    {"nome": "Pedro", "nota": 7.5},
    {"nome": "Ana",   "nota": 9.0},
]

print("📊 Classificador de Desempenho Acadêmico\n")

for aluno in alunos:
    nota = aluno["nota"]
    if nota >= 9.0:
        categoria = "Excelente 🏆"
    elif nota >= 7.5:
        categoria = "Bom 👍"
    elif nota >= 6.0:
        categoria = "Regular ⚠️"
    else:
        categoria = "Precisa melhorar ❗"
    print(f"{aluno['nome']:8} | Nota: {nota:4.1f} | {categoria}")
```
