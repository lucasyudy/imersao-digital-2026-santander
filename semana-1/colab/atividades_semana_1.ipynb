## Desafio: organizando o suporte ao cliente

Você foi contratado como analista em uma empresa de tecnologia e recebeu a seguinte demanda do time de Customer Success:
#
*“Precisamos melhorar a forma como lidamos com os pedidos de suporte dos usuários. As mensagens chegam com vários problemas misturados, como dificuldades para acessar o sistema, dúvidas sobre pagamento ou erros no uso de funcionalidades. Está tudo confuso e difícil de responder de forma ágil.”*
#
Utilizando os fundamentos do pensamento computacional, a proposta é organizar e automatizar o atendimento por meio de categorização, identificação de padrões, abstração do fluxo e definição de um algoritmo simples de resposta.

### Resposta
#
Para resolver esse problema, o primeiro passo é aplicar a **decomposição**, separando os atendimentos em categorias principais, como:
- acesso ao sistema;
- pagamento;
- uso da plataforma.
#
Essa divisão facilita a organização do suporte e permite tratar cada tipo de solicitação de forma mais objetiva.
#
Em seguida, é possível aplicar o reconhecimento de **padrões**, observando que muitas dúvidas se repetem dentro de cada categoria. Por exemplo, em acesso ao sistema podem aparecer com frequência problemas de senha, em pagamento, dúvidas sobre cobrança, e em uso da plataforma, dificuldades para localizar funcionalidades.
#
A próxima etapa é a **abstração**, criando modelos simplificados para o atendimento, como respostas padrão, perguntas frequentes (FAQ) e formulários com opções pré-definidas. Assim, em vez de analisar cada mensagem do zero, o time trabalha com uma estrutura mais clara e padronizada.
#
Por fim, é viável construir um **algoritmo** para lidar com cada solicitação. Esse fluxo pode seguir a sequência:
#
**Receber mensagem → Identificar categoria → Verificar se existe resposta padrão → Enviar resposta automática ou encaminhar para o setor responsável**.
#
Dessa forma, o atendimento se torna mais rápido, organizado e eficiente, reduzindo o tempo de resposta e melhorando a experiência do usuário.

### Exemplo de algoritmo em pseudocódigo
#
```
Início
Receber mensagem do usuário
Identificar categoria da solicitação
Se categoria tiver resposta padrão:
    enviar resposta automática
Senão:
    encaminhar para o setor responsável
Fim
```

````python
# Exemplo simples em Python inspirado na solução do desafio

def classificar_solicitacao(mensagem):
    mensagem = mensagem.lower()

    if "senha" in mensagem or "login" in mensagem or "acesso" in mensagem:
        return "acesso"
    elif "pagamento" in mensagem or "cobrança" in mensagem or "boleto" in mensagem:
        return "pagamento"
    elif "erro" in mensagem or "funcionalidade" in mensagem or "plataforma" in mensagem:
        return "uso da plataforma"
    else:
        return "outros"

def responder_suporte(mensagem):
    categoria = classificar_solicitacao(mensagem)

    respostas = {
        "acesso": "Resposta automática: orientações para redefinição de senha e recuperação de acesso.",
        "pagamento": "Resposta automática: informações sobre cobrança, boleto e confirmação de pagamento.",
        "uso da plataforma": "Resposta automática: passo a passo de uso da funcionalidade informada."
    }

    if categoria in respostas:
        return f"Categoria: {categoria}\n{respostas[categoria]}"
    else:
        return f"Categoria: {categoria}\nEncaminhar para atendimento humano."

mensagem_teste = "Não consigo fazer login na plataforma porque esqueci minha senha."
print(responder_suporte(mensagem_teste))
````

## Desafio: automatizando processos na equipe de RH
#
Você está atuando como analista em um projeto com o time de Recursos Humanos de uma empresa.
A equipe está sobrecarregada porque precisa, toda semana, organizar entrevistas com candidatos,
enviar e-mails de confirmação e montar uma agenda com todos os horários.

*“Estamos gastando muito tempo com tarefas repetitivas. Será que dá pra facilitar esse processo?”*

Utilizando a abordagem de pensamento computacional, a proposta é organizar esse fluxo
para reduzir o esforço manual e tornar o processo mais ágil.

### Resposta
#
Para resolver esse problema, o primeiro passo é aplicar a **decomposição do processo**,
dividindo a tarefa principal de organizar entrevistas em etapas menores, como:
- coletar a disponibilidade dos candidatos;
- enviar os convites;
- receber as confirmações;
- montar a agenda final.
#
Essa divisão facilita a visualização do fluxo, ajuda a identificar gargalos
e mostra com clareza quais atividades são repetitivas e podem ser automatizadas.
#
Em seguida, é importante observar os **padrões**. O processo de entrevista costuma seguir
sempre a mesma sequência: primeiro acontece o contato com o candidato, depois a resposta,
em seguida a confirmação do horário e, por fim, a reserva da agenda. Reconhecer esse padrão
permite criar soluções reutilizáveis e padronizadas para o RH.
#
Ao aplicar a **abstração**, podemos representar todo esse conjunto de etapas como um único fluxo
automatizado, utilizando por exemplo uma ferramenta de agendamento. Nesse modelo, o candidato
recebe um e-mail com link para escolher o horário disponível, e o sistema registra a escolha
em um calendário compartilhado, escondendo a complexidade operacional do processo.
#
Por fim, podemos definir um **algoritmo** para automatizar o fluxo:
**Receber a lista de candidatos → Enviar e-mail com link de agendamento → Confirmar horário escolhido → Atualizar agenda da equipe → Enviar lembrete automático antes da entrevista**.
#
Dessa forma, o trabalho do RH se torna mais organizado, rápido e escalável,
reduzindo tarefas manuais e melhorando a experiência tanto da equipe quanto dos candidatos.

### Exemplo de algoritmo em pseudocódigo
#

```
Início
Receber lista de candidatos
Para cada candidato:
    enviar e-mail com link de agendamento
    aguardar escolha do horário
    confirmar entrevista
    atualizar agenda da equipe
    enviar lembrete automático
Fim
```

````python
# Exemplo simples em Python inspirado na solução do desafio

candidatos = [
    {"nome": "Ana", "email": "ana@email.com", "horario": "10:00"},
    {"nome": "Bruno", "email": "bruno@email.com", "horario": "14:00"},
    {"nome": "Carla", "email": "carla@email.com", "horario": "16:00"}
]

def processar_entrevistas(lista_candidatos):
    agenda = []

    for candidato in lista_candidatos:
        nome = candidato["nome"]
        email = candidato["email"]
        horario = candidato["horario"]

        print(f"Enviando e-mail de agendamento para {nome} ({email})...")
        print(f"Horário confirmado: {horario}")
        print(f"Atualizando agenda da equipe com a entrevista de {nome}...")
        print(f"Enviando lembrete automático para {nome}...\n")

        agenda.append({"nome": nome, "horario": horario})

    return agenda

agenda_final = processar_entrevistas(candidatos)
agenda_final
````

A aplicação do pensamento computacional permite organizar o processo de entrevistas de forma lógica,

## Desafio: Automatizando tarefas
#
Você está trabalhando como pessoa desenvolvedora júnior em uma equipe que está criando
um sistema para auxiliar no controle de presença em eventos online. O time precisa de um
algoritmo que processe a lista de participantes e envie um e-mail de agradecimento
apenas para quem participou de toda a transmissão ao vivo.
#
> “Queremos automatizar o envio de agradecimentos. Mas só para quem assistiu à transmissão
> do início ao fim. Você consegue organizar esse processo em um algoritmo?”
#

### Solução
#
**1. Dados de entrada necessários**  
- Lista de participantes do evento.  
- Para cada participante: horário de entrada e horário de saída na transmissão.  
- Duração total do evento (em minutos, por exemplo) .
#
**2. Verificação da presença completa**  
- Para cada participante, calcular o tempo assistido:  
`tempo_assistido = horário_de_saída - horário_de_entrada.`
- Comparar esse tempo com a duração total da transmissão .
#
**3. Decisão com base nos dados**  
- Se o tempo assistido for igual ou maior que o tempo total do evento, o participante
será considerado como tendo participado de toda a transmissão .  
- Caso contrário, ele não se enquadra no critério de presença completa.
#
**4. Ação final**  
- Para cada participante com presença completa, enviar um e-mail de agradecimento.  
- Para os demais, não enviar o e-mail .
#
**Algoritmo (passo a passo)**  
1. Receber a lista de participantes com horários de entrada e saída.  
2. Para cada participante da lista:  
    a. Calcular o tempo assistido (saída - entrada).  
    b. Verificar se o tempo assistido é igual ou maior que a duração total do evento.  
    c. Se for, marcar o participante para receber o e-mail de agradecimento.  
    d. Senão, não marcar para envio.  
3. Enviar os e-mails de agradecimento para todos os participantes marcados.

### Exemplo de algoritmo em pseudocódigo

```
Início
  receber duração_total_evento
  receber lista_participantes (com horário_entrada e horário_saida)
  para cada participante em lista_participantes:
      calcular tempo_assistido = horário_saida - horário_entrada
      se tempo_assistido >= duração_total_evento:
          marcar para envio de e-mail de agradecimento
      senão:
          não enviar e-mail
  enviar e-mails para todos os participantes marcados
Fim
```

````python
# Exemplo simples em Python inspirado no algoritmo descrito

participantes = [
    {"nome": "Ana",   "entrada": 0,   "saida": 60},
    {"nome": "Bruno", "entrada": 5,   "saida": 40},
    {"nome": "Carla", "entrada": 0,   "saida": 60},
]

duracao_total_evento = 60  # em minutos

def participantes_com_presenca_completa(lista, duracao_total):
    aptos = []

    for p in lista:
        tempo_assistido = p["saida"] - p["entrada"]

        if tempo_assistido >= duracao_total:
            aptos.append(p["nome"])

    return aptos

aptos_para_agradecimento = participantes_com_presenca_completa(
    participantes,
    duracao_total_evento
)

print("Participantes com presença completa:")
for nome in aptos_para_agradecimento:
    print(f"- {nome}")

# Aqui, em um sistema real, entraria a lógica de envio de e-mail
# para cada participante listado acima.
````

### Comentário final sobre o desafio
#
Esse desafio mostra como um problema do mundo real, enviar e-mails apenas para quem
participou de toda a transmissão pode ser convertido em um algoritmo claro e executável.
A partir da definição dos dados de entrada, da verificação da condição de presença completa
e da decisão de envio ou não de e-mail, fica evidente como a lógica de programação ajuda
a automatizar tarefas repetitivas e a reduzir o trabalho manual.

## Desafio: Classificando desempenho acadêmico
#
Você está trabalhando no desenvolvimento de um sistema educacional que precisa exibir
mensagens personalizadas para estudantes com base em sua média final.
#
A regra definida pela equipe pedagógica é:
 - Média menor que 5,0: mensagem **"Você está reprovado."**
 - Média entre 5,0 e 6,9: mensagem **"Você está de recuperação."**
 - Média 7,0 ou mais: mensagem **"Parabéns! Você foi aprovado."**
#
Sua tarefa é criar um algoritmo em **linguagem natural** que represente esse processo de
verificação e decisão, usando **estruturas condicionais**.

### Solução
#
 1. Solicitar a média final do estudante.  
 2. Verificar o valor da média usando condições encadeadas:  
    - Se a média for menor que 5,0:  
      - Exibir a mensagem: **"Você está reprovado."**  
    - Senão, se a média for menor que 7,0:  
      - Exibir a mensagem: **"Você está de recuperação."**  
    - Senão:  
      - Exibir a mensagem: **"Parabéns! Você foi aprovado."**
#
Essa estrutura garante que apenas uma mensagem será exibida para cada estudante,
de forma clara e objetiva.

### Exemplo de algoritmo em pseudocódigo

```
Início  
  ler média_final  
  se média_final < 5,0 então  
      exibir "Você está reprovado."  
  senão se média_final < 7,0 então  
      exibir "Você está de recuperação."

Fim
```

de valores determinam respostas específicas.

## Desafio: Calculando pedidos
#
Você está desenvolvendo uma funcionalidade para um aplicativo de lanchonete.  
O sistema precisa calcular o valor total de um pedido com base na quantidade de itens solicitados.
#
Itens disponíveis:
 - Hambúrguer: R$ 12,00

 - Batata frita: R$ 7,00

 - Refrigerante: R$ 5,00
#
Sua tarefa é criar um algoritmo em linguagem natural (ou fluxograma) que:
 - armazene os preços e quantidades em variáveis;
 - calcule o valor total do pedido com base nas quantidades informadas.

### Solução
#
1. Criar variáveis para armazenar os preços fixos dos itens:
   - `preco_hamburguer = 12`
   - `preco_batata = 7`
   - `preco_refri = 5`
#
2. Criar variáveis para guardar as quantidades que o cliente informar:
   - `qtd_hamburguer`
   - `qtd_batata`
   - `qtd_refri`
#
3. Calcular o subtotal de cada item:
   - `total_hamburguer = qtd_hamburguer * preco_hamburguer`
   - `total_batata = qtd_batata * preco_batata`
   - `total_refri = qtd_refri * preco_refri`.
#
4. Calcular o valor total do pedido somando os subtotais:
   - `total_pedido = total_hamburguer + total_batata + total_refri`.
#
5. Exibir uma mensagem clara para o cliente informando o valor total a pagar.

### Algoritmo em pseudocódigo

```
Início  
  preco_hamburguer ← 12  
  preco_batata ← 7  
  preco_refri ← 5  
  ler qtd_hamburguer  
  ler qtd_batata  
  ler qtd_refri  
  total_hamburguer ← qtd_hamburguer * preco_hamburguer  
  total_batata ← qtd_batata * preco_batata  
  total_refri ← qtd_refri * preco_refri  
  total_pedido ← total_hamburguer + total_batata + total_refri  
  exibir "Valor total do pedido: R$", total_pedido  
# Fim
```

````python
# Exemplo em Python inspirado no algoritmo do desafio

preco_hamburguer = 12.00
preco_batata = 7.00
preco_refri = 5.00

qtd_hamburguer = int(input("Quantidade de hambúrgueres: "))
qtd_batata = int(input("Quantidade de batatas fritas: "))
qtd_refri = int(input("Quantidade de refrigerantes: "))

total_hamburguer = qtd_hamburguer * preco_hamburguer
total_batata = qtd_batata * preco_batata
total_refri = qtd_refri * preco_refri

total_pedido = total_hamburguer + total_batata + total_refri

print(f"Valor total do pedido: R$ {total_pedido:.2f}")
````

### Comentário sobre o desafio
#
Esse desafio trabalha diretamente com **variáveis numéricas** e operações de multiplicação
e soma, exatamente como é comum em sistemas de pedidos reais. Ao separar preços,
quantidades e subtotais em variáveis, o algoritmo fica mais organizado e fácil de manter,
além de deixar clara a relação entre os dados e o cálculo final apresentado ao cliente.

## Desafio: Calculando valor da entrega
#
Você está desenvolvendo um sistema para uma empresa de delivery. O valor da taxa de entrega
depende da distância até o cliente e se o pedido foi feito em um dia de chuva.
#
Regras:
- Até 5 km: taxa de R$ 5,00.
- Entre 5 e 10 km: taxa de R$ 8,00.
- Acima de 10 km: taxa de R$ 10,00.
- Se estiver chovendo, acrescentar R$ 2,00 à taxa padrão.
#
O objetivo é criar um algoritmo em linguagem natural que informe o valor final da entrega,
seguindo essas regras.

### Solução em linguagem natural
#
**1. Entrada**  
- Solicitar a distância até o endereço do cliente (em km).  
- Perguntar se está chovendo (resposta booleana: verdadeiro/falso ou sim/não).
#
**2. Processamento**  
- Verificar a distância informada e definir a taxa base:
  - Se a distância for menor ou igual a 5 km, taxa base = R$ 5,00.

  - Se a distância for maior que 5 km e menor ou igual a 10 km, taxa base = R$ 8,00.

  - Se a distância for maior que 10 km, taxa base = R$ 10,00.

- Verificar se está chovendo:

  - Se estiver chovendo, somar R$ 2,00 à taxa base.
  
- Calcular o valor final da entrega a partir da taxa base (com ou sem acréscimo).
#
**3. Saída**  
- Exibir o valor total da taxa de entrega com uma mensagem clara, por exemplo:
  - "Taxa de entrega: R$ X,00".

### Exemplo de algoritmo em pseudocódigo

```
 Início  
   ler distancia_km  
   ler chovendo (verdadeiro/falso ou sim/não)  

   se distancia_km ≤ 5 então  
       taxa_base ← 5  
   senão se distancia_km ≤ 10 então  
       taxa_base ← 8  
   senão  
       taxa_base ← 10  

   se chovendo é verdadeiro então  
       taxa_final ← taxa_base + 2  
   senão  
       taxa_final ← taxa_base  

   exibir "Taxa de entrega: R$ ", taxa_final  
 Fim
 ```

````python
# Exemplo em Python inspirado no algoritmo do desafio

distancia_km = float(input("Informe a distância até o cliente (em km): "))
resposta_chuva = input("Está chovendo? (s/n): ").strip().lower()

chovendo = resposta_chuva == "s"

if distancia_km <= 5:
    taxa_base = 5.00
elif distancia_km <= 10:
    taxa_base = 8.00
else:
    taxa_base = 10.00

if chovendo:
    taxa_final = taxa_base + 2.00
else:
    taxa_final = taxa_base

print(f"Taxa de entrega: R$ {taxa_final:.2f}")
````

###  Comentário sobre o desafio

 Esse desafio combina **condicionais encadeadas** com uma verificação booleana adicional,
 mostrando como regras de negócio reais podem depender tanto de valores numéricos
 (distância em km) quanto de condições lógicas simples (está chovendo ou não).
 A estrutura em etapas de **entrada → processamento → saída** ajuda a deixar o algoritmo
 mais organizado e mais fácil de implementar em qualquer linguagem de programação.

A tarefa é criar um algoritmo em linguagem natural que use uma **função** para fazer

e retornar o valor convertido.

### Solução em linguagem natural
#

   - `taxa_de_cambio`.
#
2. Dentro da função, calcular o valor convertido usando a fórmula:

   - Solicitar ao usuário o valor em reais que deseja converter.
   - Solicitar a taxa de câmbio atual definida pela empresa.
   - Chamar a função `converter_para_dolar` passando esses dois valores.
   - Exibir o resultado com uma mensagem do tipo:
     - "Com a taxa atual, o valor em dólares é: [valor]".
#
Essa abordagem organiza o código em um bloco reutilizável (a função), o que facilita
futuras alterações, como adicionar novas moedas ou ajustar a fórmula de conversão.

### Exemplo de algoritmo em pseudocódigo
```
 Função converter_para_dolar(valor_em_reais, taxa_de_cambio)  
   valor_convertido ← valor_em_reais / taxa_de_cambio  
   retornar valor_convertido  

 Início (programa principal)  
   ler valor_em_reais  
   ler taxa_de_cambio  
   resultado ← converter_para_dolar(valor_em_reais, taxa_de_cambio)  
   exibir "Com a taxa atual, o valor em dólares é: ", resultado  
 Fim
 ```

````python
# Exemplo em Python inspirado no algoritmo do desafio

def converter_para_dolar(valor_em_reais, taxa_de_cambio):
    return valor_em_reais / taxa_de_cambio

valor_em_reais = float(input("Informe o valor em reais (R$): "))
taxa_de_cambio = float(input("Informe a taxa de câmbio (R$ por US$): "))

valor_convertido = converter_para_dolar(valor_em_reais, taxa_de_cambio)

print(f"Com a taxa atual, o valor em dólares é: US$ {valor_convertido:.2f}")
````

## Desafio: Simulador de pedidos em uma cafeteria
#
Você foi contratado por uma cafeteria que deseja automatizar o atendimento no balcão.
O sistema deve permitir que o atendente registre os pedidos de cada cliente, calcule
o valor total e aplique um **desconto de 10% para clientes cadastrados**.
#
O processo funciona assim:
- O atendente informa quantos itens o cliente vai pedir.
- Para cada item, o sistema solicita o nome e o preço.
- Ao final, o sistema pergunta se o cliente é cadastrado.
- Se for, aplica o desconto e exibe o valor com desconto.
- Caso contrário, exibe o valor cheio.
#
O desafio é criar um algoritmo que represente essa lógica de forma completa.

### Solução

**1. Entrada**  
- Pedir a quantidade de itens que o cliente vai pedir.  
- Criar uma variável `total` iniciando em 0.  
- Para cada item:
  - Solicitar o nome do item.
  - Solicitar o preço do item.
  - Somar o preço do item ao `total`.
#
**2. Processamento**  
- Ao final da leitura dos itens, perguntar se o cliente é cadastrado
  (pode ser `True/False` ou “sim/não”).  
- Se o cliente for cadastrado:
  - Calcular `desconto = total * 0.10`.  
  - Calcular `total_final = total - desconto`.  
- Senão:
  - `total_final = total`.
#
**3. Saída**  
- Exibir o valor total a ser pago pelo cliente, deixando claro se houve desconto
  ou se o valor é cheio.

### Exemplo de algoritmo em pseudocódigo
```
 Início  
   ler quantidade_itens  
   total ← 0  

   para i de 1 até quantidade_itens faça  
       ler nome_item  
       ler preco_item  
       total ← total + preco_item  
   fim-para  

   ler cliente_cadastrado (verdadeiro/falso ou sim/não)  

   se cliente_cadastrado é verdadeiro então  
       desconto ← total * 0.10  
       total_final ← total - desconto  
   senão  
       total_final ← total  

   exibir "Valor total a pagar: R$ ", total_final  
 Fim
 ```

````python
# Exemplo em Python inspirado no algoritmo do desafio

quantidade_itens = int(input("Quantos itens o cliente vai pedir? "))

total = 0.0

for i in range(1, quantidade_itens + 1):
    nome_item = input(f"Nome do item {i}: ")
    preco_item = float(input(f"Preço de '{nome_item}': R$ "))
    total += preco_item

resposta_cadastrado = input("O cliente é cadastrado? (s/n): ").strip().lower()
cliente_cadastrado = resposta_cadastrado == "s"

if cliente_cadastrado:
    desconto = total * 0.10
    total_final = total - desconto
else:
    desconto = 0.0
    total_final = total

print(f"\nTotal dos itens: R$ {total:.2f}")
print(f"Desconto aplicado: R$ {desconto:.2f}")
print(f"Valor final a pagar: R$ {total_final:.2f}")
````

### Comentário sobre o desafio

Esse desafio integra vários conceitos trabalhados no curso: entrada de dados em laço
(vários itens), acumulação de total, uso de condição booleana para clientes cadastrados
e cálculo de desconto proporcional ao valor do pedido. Ele se aproxima bastante
de um cenário real de sistemas de caixa, mostrando como algoritmos simples podem apoiar
diretamente processos de atendimento.
